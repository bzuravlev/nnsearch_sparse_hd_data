```python
def sequential_cos_sim(db, q):
    #for x in range(db.shape[0]):
        #for y in range(q.shape[0]):
            #sim = skmet.pairwise.cosine_similarity(x,y)
    return skmet.pairwise.cosine_similarity(db,q)
```

```python
def print_gs_matrix(matrix):
    plt.imshow(matrix, cmap='gray')
    plt.colorbar()
    plt.show()
```

```python
def calc_hard_matrix(matrix):
    # Find the indices of the maximum values in each row
    max_indices = np.argmax(matrix, axis=1)

    # Create a new matrix with the same shape, filled with zeros
    result = np.zeros_like(matrix)

    # Set the maximum values to 1
    result[np.arange(matrix.shape[0]), max_indices] = 1
    return result
```

```python
res = sequential_cos_sim(nl_spring_seer, nl_winter_seer)
```

```python
print_gs_matrix(res)
```

```python
print_gs_matrix(calc_hard_matrix(res))
```

*Number of entries per Dimension*

```python
def get_entries_per_dimension(sparse_mat):
    num = sparse_mat.shape[0]
    res = np.zeros(num)
    for i in range(num):
        res[i] = sp.csr_matrix.count_nonzero(sparse_mat[:,i])
    return res
```

```python
entries_per_dimension_garden_ds = get_entries_per_dimension(garden_ds_seer)
entries_per_dimension_garden_q = get_entries_per_dimension(garden_q_seer)
entries_per_dimension_nl_spring = get_entries_per_dimension(nl_spring_seer)
entries_per_dimension_nl_winter = get_entries_per_dimension(nl_winter_seer)
```

```python
x = np.arange(200)

plt.bar(x - 0.2, entries_per_dimension_garden_ds, 0.4, label = 'Garden Descriptors') 
plt.bar(x + 0.2, entries_per_dimension_garden_q, 0.4, label = 'Garden Query') 

#plt.xticks(x) 
plt.xlabel("Dimension") 
plt.ylabel("Number of Entries") 
plt.title("Number of Entries in each Dimension") 
plt.legend() 
plt.rcParams["figure.figsize"] = (10,3)
plt.show() 
```
